```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import OneHotEncoder
```

```python
# Read the data into a Pandas DataFrame
mental_df = pd.read_excel('mental.xlsx')
mental_df.head(30)
```

```python
# Get a brief summary of the DataFrame.
mental_df.info()
```

```python
missing_data = mental_df.isnull().sum()
missing_data
```

```python
mental_df['Suicidal thoughts'].value_counts()
mental_df['Suicidal thoughts'] = mental_df['Suicidal thoughts'].replace('YES ', "YES")
mental_df['Suicidal thoughts'].value_counts()
```

```python
mental_df = mental_df.drop('Patient Number', axis = 1)
mental_df.head()
```

```python
yes_no_cols = ['Mood Swing', 'Suicidal thoughts', 'Anorxia', 'Authority Respect', 'Try-Explanation', 'Aggressive Response', 'Ignore & Move-On', 'Nervous Break-down', 'Admit Mistakes', 'Overthinking']
for col in yes_no_cols:
    mental_df[col] = mental_df[col].map({'YES': 1, 'NO': 0}).astype(int)
mental_df.head()
```

```python
frequency_columns = ['Sadness', 'Euphoric', 'Exhausted', 'Sleep dissorder']
for column in frequency_columns:
    mental_df[column] = mental_df[column].map({'Seldom': 0, 'Sometimes': 1, 'Usually': 2, 'Most-Often': 3}).astype(int)

mental_df.head()
```

```python
from_columns = ['Sexual Activity', 'Concentration', 'Optimisim']
for column in from_columns:
    mental_df[column] = mental_df[column].astype(str).str.extract('(\d)')
    mental_df[column] = pd.to_numeric(mental_df[column])

mental_df.head()
```

```python
X = mental_df.drop(columns=['Expert Diagnose'],axis=1)
y = mental_df['Expert Diagnose']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)
```

```python


X_test = X_test.replace('Usually', np.nan)
X_test.dropna(inplace=True)
log_reg = LogisticRegression(max_iter = 1000)
log_reg.fit(X_train, y_train)

y_pred_test = log_reg.predict(X_test)
y_pred_train = log_reg.predict(X_train)

# evaluate the model
# calculate accuracy on the training set
train_accuracy = accuracy_score(y_train, y_pred_train)
print(f"Training Accuracy: {train_accuracy:.2%}")

# calculate accuracy on the test set
lr_accuracy = accuracy_score(y_test, y_pred_test)
print(f"Testing Accuracy: {lr_accuracy:.2%}")
```
